```python
import pandas as pd
from sklearn import preprocessing
import math
```

```python
df = pd.read_csv('../data/weather_baseball.csv')
df
```

```python
def norm(df, col):
    # Create x, where x the 'scores' column's values as floats
    x = df[[col]].values.astype(float)

    # Create a minimum and maximum processor object
    min_max_scaler = preprocessing.MinMaxScaler()

    # Create an object to transform the data to fit minmax processor
    x_scaled = min_max_scaler.fit_transform(x)

    # Run the normalizer on the dataframe
    df[col] = x_scaled
    return df
```

```python
df = norm(df, 'Temperature')
df = norm(df, 'Humidity')
df
```

```python
df['Sunny'] = pd.np.where(df['Outlook'] == 'sunny',1,0)
df['Overcast'] = pd.np.where(df['Outlook'] == 'overcast',1,0)
df['Rainy'] = pd.np.where(df['Outlook'] == 'rainy',1,0)
df['Windy'] = pd.np.where(df['Windy'] == True,1,0)
```

```python
df
```

```python
def distances(df, query):
    for idx, *row in df.itertuples():
        distance = 0
        for qidx, col in enumerate(row):
            distance += (col - query[qidx])**2
        distance = math.sqrt(distance)
        print(idx, distance)
```
