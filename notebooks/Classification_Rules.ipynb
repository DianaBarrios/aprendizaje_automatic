```python
def PRISM(data, class_attr):
    for class_value in data[class_attr].unique():
        E = data.copy()
        while len(E[E[class_attr] == class_value]) > 0:
            I = E.copy()
            rule = []
            attributes = [col for col in data.columns if col != class_attr]
            best_purity = 0
            best_p = 0
            while len(attributes) != 0 and best_purity != 1:
                best_purity = 0
                best_p = 0
                rule_to_add = None
                for attr in attributes:
                    for value in I[attr].unique():
                        covered = I[I[attr] == value]
                        t = len(covered)
                        satisfied = covered[covered[class_attr] == class_value]
                        p = len(satisfied)
                        purity = p/t
                        if (purity, p) > (best_purity, best_p):
                            best_purity, best_p = purity, p
                            rule_to_add = (attr, value)
                rule.append(rule_to_add)
                attributes.remove(rule_to_add[0])
                I = I[I[rule_to_add[0]] == rule_to_add[1]]
            print('If ' + ' and '.join([f"{atr} = {val}" for atr, val in rule]) + f' then {class_attr} = {class_value}\n')
            satisfied_query = ' and '.join([f'{atr} == "{val}"' for atr, val in rule])
            E = E.query(f"not({satisfied_query})")
```

```python
PRISM(df, 'recommendation')
```
